```python
df2 = pd.read_csv('../data/geospatial/country_coordinates.csv')
```

```python
df2 = df2[['Alpha-3 code', 'Latitude (average)', 'Longitude (average)']]
```

```python
df2 = df2.rename(columns = {'Alpha-3 code': 'ISO3_code', 'Latitude (average)': 'Latitude', 'Longitude (average)': 'Longitude'})
```

```python
df2['ISO3_code']=df2['ISO3_code'].str.split('"', expand=True).drop(columns=[0,2])
```

```python
df2['Latitude']=df2['Latitude'].str.split('"', expand=True).drop(columns=[0,2])
```

```python
df2['Longitude']=df2['Longitude'].str.split('"', expand=True).drop(columns=[0,2])
```

```python
df2
```

```python
df3 = pd.merge(df_splice, df2, left_on='ISO3_code', right_on='ISO3_code', how = 'left')
```
